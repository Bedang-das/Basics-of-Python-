```python
#Write a programe to check if a Date is valid or not....also print the next date

date = input("Enter the date in format {dd/mm/yyyy} ")

dd, mm, yy = date.split("/")
dd = int(dd)
mm = int(mm)
yy = int(yy)

if mm < 1 or mm > 12:
    print("Date is invalid.")

else:
    if mm in (1, 3, 5, 7, 8, 10, 12):
        max1 = 31
    elif mm in (4, 6, 9, 11):
        max1 = 30
    else:  # February
        if (yy % 4 == 0 and yy % 100 != 0) or (yy % 400 == 0):
            max1 = 29
        else:
            max1 = 28
    if dd < 1 or dd > max1:
        print("Date is invalid.")

    else:
        # Increment date
        if dd == max1 and mm != 12:
            dd = 1
            mm += 1
        elif dd == 31 and mm == 12:
            dd = 1
            mm = 1
            yy += 1
        else:
            dd += 1

        print(f"The incremented date is: {dd:02d}/{mm:02d}/{yy}")
```

```python
''' Problem Statement - 8   -   Write a Python program to calculate the gross salary
of an employee based on basic salary.'''

basic_sal = int(input("Enter the Basic Salary: "))

if(basic_sal <= 10000):
  HRA = (20/100) * basic_sal
  DA = (80/100) * basic_sal

elif(basic_sal <= 20000):
  HRA = (25/100) * basic_sal
  DA = (90/100) * basic_sal

else:
  HRA = (30/100) * basic_sal
  DA = (95/100) * basic_sal

gross_sal = basic_sal + HRA + DA

print("The Gross Salary is: ", gross_sal)
```

```python
''' Problem Statement - 9   -   Write a Python program to calculate the income
tax based on annual income'''

salary = int(input("Enter the salary: "))

if(salary <= 250000):
  tax = 0
elif(salary <= 500000):
  tax = (salary - 250000) * 0.05

elif(salary <= 1000000):
  tax = (250000 * 0.05) + (500000 * 0.20) + (salary - 1000000) * 0.30

print("The Tax amount to be paid is ", tax)
print("The salary In - Hand is ", salary - tax)
```
