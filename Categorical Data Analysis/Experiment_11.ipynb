```python
import pandas as pd
data = {
    "order_id":["Electronics","Clothing","Grocery","Clothing","Electronics","Clothing","Grocery"],
    "Payment_type":["UPI","cash","NEFT","Card","UPI","cash","UPI"],
    "Delivery_type":["Express","Standard","Express","Standard","Express","Standard","express"],
    "Customer_type":["new","repeat","new","repeat","repeat","repeat","new"]
}
df = pd.DataFrame(data)
print(df)
```

```python
#Frequency Count
# Count how many orders in each categories
frequency_count = df["order_id"].value_counts()
print(frequency_count)
```

```python
#Unique Values
#Display all the unique values
print(df["order_id"].unique())
```

```python
#Number of unique values
print(df["order_id"].nunique())

```

```python
#Frequency of the payment methods
print(df['Payment_type'].value_counts())

#Write the method using which maximum payment is
print("The most popular payment technique is",max(df["Payment_type"].value_counts()))
```

```python
#Cross tabulation of delivery vs customer type
print(pd.crosstab(df["Delivery_type"],df["Customer_type"]))
```

```python
#PErcentage Distribution
print(df["Delivery_type"].value_counts(normalize=True)*100)
```

```python
#Filtering Data
electronics_order= df[df["order_id"]=="Electronics"]
print(electronics_order)

Clothing_data = df[df["order_id"]=="Clothing"]
print(Clothing_data)

Grocery_data= df[df["order_id"]=="Grocery"]
print(Grocery_data)


```

```python
#grouping
#Order by category and payment method

print(df.groupby("order_id")["Payment_type"].value_counts())
```

```python
#Sorting Data
sorted_df = df.sort_values(by="order_id")
print(sorted_df)
```
