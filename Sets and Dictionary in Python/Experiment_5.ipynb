```python
#Creation of sets
set1 = {"Apple","oranges","banana","pajamas","cars","helicopters","popper"}
print(set1)

#duplicates are not allowed
set1 = {"Apple","oranges","banana","pajamas","cars","helicopters","popper","Apple"}
print(set1)

#The value True and 1 are considered the same calue in set, and are the
thisset = {"Apple","oranges","banana","pajamas","cars","helicopters","popper",False,0,1,2}
print(thisset)

```

```python
#Set allows all types of data set
set1 = {"pop",10.54,True,34}
print(set1)
```

```python
#CHEcking the set
seet = {"Apple","oranges","banana","pajamas","cars","helicopters","popper"}
print("banana" in seet)
if "banana" in seet:
  print("Yes,'banana' is in the set")
if "mango" not in  seet:
  print("No,'mango' is not in the set")
```

```python
#In set we cannot replace items however we can add new items
set1 = {"Apple","oranges","banana","pajamas","cars","helicopters","popper"}
##set1[1] = "mango"
##print(set1)

#NOW TO ADD
set1.add("mango")
print(set1)

#Remove items
if "pajamas" in set1:
  print("Yes,'pajamas' is in the set, so we are removing it ")
  set1.remove("pajamas")
  print(set1)

```

```python
#Set operations on set
set1 = {"Apple","oranges","banana","pajamas","cars","helicopters","popper"}
set2 = {"Apple","oranges","banana","pajamas","cars","helicopters","popper","mango"}

#Union
print("Union:",set1|set2)

#Intersection
print("Intersection:",set1&set2)

#Difference
print("Difference:",set1-set2)

#Symmetric Difference
print("Symmetric Difference:",set1^set2)
```

```python
#Frozen-set
check_1 = frozenset(["Apple","oranges","banana","pajamas","cars","helicopters","popper"])
print(check_1)
print(type(check_1))
```

```python
#Adding or removal of elements in frozen-set(not possible)
check_1 = frozenset(["Apple","oranges","banana","pajamas","cars","helicopters","popper"])
check_1.add("mango")
```

```python
#how to add elements in frozen-set
check_1=frozenset({"Apple","oranges","banana","pajamas","cars","helicopters","popper"})
check_2 = frozenset(["mango"])
print("Union:",check_1|check_2)
```

```python
''' Problem Statement - 2 - Write a Python Program to search for a student's marks using the name
entered, if the student is not found, Display a suitable message'''

marks = {"Lehalyaa" : 90, "Aditi" : 95, "Gayatri" :90, "Bedang" : 93, "Arjit" : 65, }
name = input("Enter the name of the student: ")

if name in marks:
    print("The marks of", name, "are", marks[name])
else:
    print("The student name isn't present in the Directory")
```

```python
# Problem statement - 2  using get() function

name = input("Enter the name of the student: ")
print(marks.get(name, "Student Name not found in the Directory"))
```

```python
''' Problem Statement - 3 - A Ditionary stores usernames and passwords. Write a
Python program to validate user login'''

user_pass = {"Lehalyaa" : 123456, "Aditi0" : "cutie_pie", "Gayatri" : "ping_pong", "Arjit" : 98416647}
user = input("Enter the username: ")
password = input("Enter the password: ")

if user in user_pass:
  if user_pass[user] == str(password) or user_pass[user] == int(password):
    print("\n Username and Password are verified, Login Successful")
  else:
    print("\n Password is Incorrect, Re- enter")
else:
  print("\n Both Username and password are incorrect")
```

```python
''' Problem Statement - 4 - A Dictionary stores student names and marks. Write a
Python program to find the student who scored the highest marks'''

marks = {"Lehalyaa" : 90, "Aditi" : 95, "Gayatri" :90, "Bedang" : 93, "Arjit" : 65, }
max_stu = max(marks, key = marks.get)
print("Student with Highest marks: ", max_stu)
print("The highest marks are: ", marks[max_stu])
```
