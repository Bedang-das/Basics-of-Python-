```python
# Tokenization
!pip install nltk

import nltk
nltk.download("punkt_tab")
nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")
```

```python
from nltk.tokenize import word_tokenize
text = "This is a sample sentence for tokenization."
tokens = word_tokenize(text)
print(tokens)
```

```python
#Sentence tokenization
from nltk.tokenize import sent_tokenize
text = "This is a sample sentence. It contains multiple sentences."
sentences = sent_tokenize(text)
print(sentences)

#Whenever there is a "." from there it will seperate
```

```python
#Paragrapgh tokenization
from nltk.tokenize import LineTokenizer
text = "This is a sample paragraph. This is the second line of para one. look carefully it does not separeate on the basis on fullstop but on the basis of gap \nIt contains multiple sentences. "
tokenizer = LineTokenizer(blanklines='keep')
print(tokenizer.tokenize(text))

#
```

```python
# Stop Word removal
from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
words = word_tokenize(text)
filtered_word = [ w for w in words if w.lower() not in stop_words]
print(filtered_word)
```

```python
#Stemming
#Stemming reduce words to their root form
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
words = ["playing", "played", "play","going","college","water","eating","hopping"]
for w in words:
  print(stemmer.stem(w))
```

```python
#lemmatization
#Lemmatization converts words back to dictionary form
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
words = ["playing", "played", "play","going","college","water","eating","hopping"]
for w in words:
  print(lemmatizer.lemmatize(w))
```

```python
#Part-of-speech (pos) tagging
#POS  tagging identifies grammatical roles of words
from nltk import pos_tag
nltk.download("averaged_perceptron_tagger_eng")
words = word_tokenize("This is the best way to tokenize this so that we are able to learn NLP ")
pos_tags = pos_tag(words)
print(pos_tags)
```

```python
from nltk.probability import FreqDist
words = word_tokenize(text)
fd = FreqDist(words)
print(fd.most_common(5))
```
