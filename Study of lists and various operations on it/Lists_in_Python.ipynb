```python
student_mark = [12,23,24,15,15,67,53,87,34,86,43,67,24]
print("The highest marks scored is:",max(student_mark))
print("The lowest marks scored is:",min(student_mark))
print("The average marks scored is:",(sum(student_mark))/len(student_mark))

print("The total marks in ascending order is:",sorted(student_mark))

student_mark.sort()
print("The total marks in ascending order is ",student_mark)

print("The total marks in descending order is:",sorted(student_mark,reverse=True))
```
